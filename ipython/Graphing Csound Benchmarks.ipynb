```python
import cPickle as pickle
```

```python
with open('../build_data-21-05-14-22_55_21.p', 'rb') as fp:
    build_data = pickle.load(fp)
```

```python
print build_data['host']
```

```python
print build_data['start_time']
```

```python
values = []
labels = []
for data in build_data['data']:
    csound_version = data['tag']
    values.append(data['time'])
    labels.append(csound_version)
bar(arange(len(values)), values)
xticks(arange(len(values)) + 0.5, labels, rotation='vertical');

title("Build times")
ylabel("Measured time (s)")
grid()
```

```python
with open('../run_data-21-05-14-22_55_21.p', 'rb') as fp:
    run_data = pickle.load(fp)
```

```python
values = []
labels = []
for data in run_data['data']:
    csound_version = data['tag']
    for tests in data['data']:
        values.append(tests['times'])
        labels.append(csound_version + ' ' + tests['csd'])
```

```python
boxplot(values);
xticks(arange(len(values)) + 1, labels, rotation='vertical');
grid()
gcf().set_figwidth(16)
title("Running times (num_runs = %i)"%len(values[0]) )
ylabel("Measured time (s)")
```
